```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import urllib.request
```

```python
def extract_link(url,bg):
  """ 
  Extraction des liens ....."
  parameters:
  url : link page....
  """
  page=urllib.request.urlopen(url,timeout=5)
  soup=bs(page)
  liste=[]
  for link in soup.findAll('a'):
      linkfound = link.get('href')
      links=('https://www.adresses-mairies.fr' + linkfound)
      liste.append(links)

  liste=list(set([l for l in liste if bg in l]))
  return liste, soup

```
