```python
def data_cleaning(annee):
    columns_save=['Country Name','Indicator Name',annee]
    new_WDIData=WDIData[columns_save]
    new_WDIData1=new_WDIData.pivot_table(index="Country Name",columns="Indicator Name",values=annee,dropna=False)
    res=(100*new_WDIData1.isnull().sum()/new_WDIData1.shape[0]).sort_values(ascending=False)
    new_WDIData2=new_WDIData1.loc[:,res<5]
    print(len(res))
    new_WDIData2_clean=new_WDIData2.dropna()
    return new_WDIData2_clean
    
```

```python
data=data_cleaning("2018")
```
